```python
df = pd.read_csv("data.csv")

# Remove extra spaces from column names
df.columns = df.columns.str.strip()

print(df.columns)
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df['Score'] = df['Score'].fillna(df['Score'].mean())

print("\nAfter Filling Missing Values:\n")
print(df)
```

## Reflection

Imputation of missing values with mean rather than zero is preferable because zero is an unreasonable figure for variables such as age or scores.

Imputation with zero can alter the dataset and have an adverse effect on the model’s training phase.

Mean imputation results in a fair estimate and also maintains the integrity of the overall dataset.

Deleting rows from the dataset is undesirable, particularly in a small dataset, and imputation should be preferred instead.
