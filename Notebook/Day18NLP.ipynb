```python
import nltk
import string

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from textblob import TextBlob

# Download required datasets
nltk.download('punkt')
nltk.download('punkt_tab')   # important fix
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
def clean_bio(text):
    # Lowercase
    text = text.lower()
    
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    # Tokenize
    tokens = word_tokenize(text)
    
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    filtered = [w for w in tokens if w not in stop_words]
    
    return " ".join(filtered)
```

```python
sample = "I love Hiking in the mountains and Coding late at night!"
print("Original:", sample)
print("Cleaned:", clean_bio(sample))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

bios = [
    "Expert in Python and Machine Learning for social good.",
    "Professional Chef who loves outdoor Hiking and mountains.",
    "Machine Learning enthusiast and mountain hiker."
]

vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(bios)

print("Vocabulary:", vectorizer.get_feature_names_out())
print("Matrix Shape:", tfidf_matrix.toarray().shape)
```

```python
bios = [
    "I absolutely love this event, it was amazing!",
    "The experience was okay, nothing special.",
    "I hated the event, it was terrible.",
    "Great organization and fun activities!",
    "Not good, very disappointing experience."
]

for bio in bios:
    analysis = TextBlob(bio)
    print("Text:", bio)
    print("Polarity:", analysis.sentiment.polarity)
    print("Subjectivity:", analysis.sentiment.subjectivity)
    print("------")
```

### Stemming vs Lemmatization

Stemming removes word endings and may produce incomplete words (e.g., "running" → "run", "studies" → "studi").
Lemmatization converts words into their meaningful base form using grammar (e.g., "running" → "run", "better" → "good").
Lemmatization is more accurate and preferred in NLP tasks.
