# Section 4: Understanding Vectorization

Vectorization refers to carrying out computations using entire arrays or datasets rather than individual elements through looping structures.
Traditional computation involves working on each element sequentially through looping, which is not efficient.
With modern computing software such as NumPy, computations are performed directly on the whole array.
This makes the computation faster since they are done at lower levels with highly optimized C codes.
For instance, instead of adding elements in an array using loops, NumPy enables us to do this operation in one go.
This is critical in AI/ML applications since the models operate on vast amounts of data.
