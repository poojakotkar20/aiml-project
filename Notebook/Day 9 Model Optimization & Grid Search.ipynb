```python
import pandas as pd
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score

# 1. Load dataset
data = fetch_california_housing()

# 2. Split data
X_train, X_test, y_train, y_test = train_test_split(
    data.data, data.target, test_size=0.2, random_state=42
)

# 3. Scale data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)   # IMPORTANT

# 4. Grid Search setup
param_grid = {'alpha': [0.1, 1.0, 10.0, 100.0, 500.0]}

grid_search = GridSearchCV(
    Ridge(),
    param_grid,
    cv=5,
    scoring='r2'
)

# 5. Run Grid Search
grid_search.fit(X_train_scaled, y_train)

print("Best Alpha Found:", grid_search.best_params_)
print(f"Best Cross-Validation R2 Score: {grid_search.best_score_:.4f}")

# 6. Default Model
ridge_default = Ridge()
ridge_default.fit(X_train_scaled, y_train)

pred_default = ridge_default.predict(X_test_scaled)
r2_default = r2_score(y_test, pred_default)

# 7. Optimized Model
best_model = grid_search.best_estimator_

pred_best = best_model.predict(X_test_scaled)
r2_best = r2_score(y_test, pred_best)

# 8. Comparison Table
comparison = pd.DataFrame({
    "Model": ["Default Ridge", "Optimized Ridge"],
    "R2 Score": [r2_default, r2_best]
})

print("\nModel Comparison:\n")
print(comparison)
```
